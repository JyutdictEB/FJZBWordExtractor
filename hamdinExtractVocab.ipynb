```python
import extractor 
import re
```

```python
rawLines = extractor.loadRawCSV('vocab.csv',";")
extracted_word = []
for line in rawLines:
  if len(line) < 4:
    continue
  des = line[3].strip('"')
  word = line[0].strip('"')
  word_pri = line[1].strip('"')
  rWord = ""
  if len(word_pri) > 0:
    rWord = word_pri
  else:
    rWord = word
  word_in_line = extractor.splitWordHamdinVocab(des)
  if len(word_in_line) == 0:
    continue
  extracted_word.append([re.sub(r"[~～]",rWord,w) for w in word_in_line])
```

```python
for w in extracted_word:
  print("\n".join(w))
```
