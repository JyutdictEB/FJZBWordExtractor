```python
import extractor 
import re
```

```python
extractor.splitWordHamdinSinglezi("①指面积、体积、容量、数量、强度、力量超过一般或超过所比较的对象，与“小”相对；②规模广，程度深，性质重要：～局；③用于“不”后，表示程度浅或次数少：不～高兴；④排行第一：老～；⑤敬辞：～作；⑥时间更远：～前年；⑦（～夫）古代官职，位于“卿”之下，“士”之上；⑧不很详细，不很准确：～概；⑨[钦]牌戏术语，自己的牌小于对方的牌出不了：三条烟？～了。")
```

```python
rawLines = extractor.loadRawCSV('singleZi.csv',";")
extracted_word = []
for line in rawLines:
  if len(line) < 4:
    continue
  des = line[3].strip('"')
  word = line[0].strip('"')
  word_pri = line[1].strip('"')
  rWord = ""
  if len(word_pri) > 0:
    rWord = word_pri
  else:
    rWord = word
  word_in_line = extractor.splitWordHamdinSinglezi(des)
  if len(word_in_line) == 0:
    continue
  extracted_word.append([re.sub(r"[~～]",rWord,w) for w in word_in_line])
```

```python
for w in extracted_word:
  print("\n".join(set(w)))
```
